```python
import pandas as pd
import numpy as np

dogs=pd.read_csv('dog.csv')
dogs.head()
```

```python
dogs_df=dogs[:20]
```

```python
def age_categorized(color):
    if color=='Black':
        return 20
    elif color=='White':
        return 35
    elif color=='Gold':
        return 32
    elif color=='Brown':
        return 25
    elif color=='Tan':
        return 23
    elif color=='Blenheim':
        return 33
    else:
        return 30
```

```python
dogs_df.loc[:,'Age']=dogs_df['animalcolour'].apply(age_categorized)
```

```python
dogs_df
```

```python
dogs_df_df=dogs_df.pivot_table(values='Age',index='animalbreed',columns='animalcolour',fill_value=0)
```
