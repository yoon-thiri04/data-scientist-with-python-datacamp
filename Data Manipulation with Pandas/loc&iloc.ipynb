### Slicing inner index levels

```python
dogs_srt2=dogs[:15].set_index(['animalcolour','animalgender']).sort_index(level=['animalcolour'])
```

```python
dogs_srt2
```

```python
dogs_srt2.loc[['Black','Brown'],['AnimalName']]
```

```python
dogs_srt2.loc['Black':'Tan',['AnimalName','animalbreed']]
```

```python
dogs_srt2.value_counts('animalbreed',sort=True)
```

```python
dogs_srt2['Age']=dogs_srt2['animalbreed'].apply(lambda x: 30 if x=='Maltese' else 40)
```

```python
dogs_srt2.reset_index()
```

```python
dogs_srt2.groupby(['animalcolour','animalgender'])['Age'].agg(np.median)
```

```python
dogs_srt2.reset_index().pivot_table(values='Age', index='animalcolour', columns='animalgender', fill_value=0 , margins=True, aggfunc=[np.median, np.std])
```

```python
dogs_srt2.loc[[('Tan','M'),('White','F')],['AnimalName']]
```

```python
dogs_srt2.loc[[('Black','F'),('Tan','F')],['AnimalName']]
```

```python
dogs_srt2.loc['Black':'Tan',['AnimalName','animalbreed']]
```

```python
dogs_srt2.loc[('Black','M'):('White','F'),['AnimalName']]
```

```python
dogs_srt2.loc[("Black",'F'):('Tan','M'),'AnimalName':'animalbreed'] # Slice Twice --> No bracket needed
```

### Using iloc()

```python
dogs_df=dogs[:15]
```

```python
dogs_df.iloc[[0,5],[0,2]] # Manually
```

```python
dogs_df.iloc[0:5,[0,2]] # Slcing the row numbers
```

```python
dogs_df.iloc[0:3,0:5] # Sling both row and col numbers
```

```python
dogs_df.iloc[:,:]
```

```python
dogs_df.loc[:,:]
```
