```python
def generate_time_slots():
    time_slots = []
    start_hour, start_minute = 9, 0  # Start at 9:00 AM
    duration = 1  # Each slot is 1 hour

    for i in range(18):
        end_hour = start_hour + (start_minute + 60 * duration) // 60
        end_minute = (start_minute + 60 * duration) % 60
        time_slots.append((f"{start_hour:02}:{start_minute:02}", f"{end_hour:02}:{end_minute:02}"))
        start_hour = end_hour
        start_minute = end_minute
    return time_slots

time_slots = generate_time_slots()
```
