```python
import pandas as pd
import matplotlib.pyplot as plt

dogs=pd.read_csv('dog.csv')
dogs.head()
```

```python
dog_df=dogs[:200]
dog_df['animalcolour'].value_counts()
```

```python
def age_categorized(color):
    if color=='Black' or color=='Grey':
        return 20
    elif color=='Tan' or color=='Wheaten' or color=='Fawn':
        return 23
    elif color=='White' or color=='Cream' or color=='Ruby':
        return 25
    elif color=='Gold' or color=='Brindle' or color=='Salt':
        return 28
    elif color=='Brown' or color=='Caramel' or color=='Blue':
        return 30
    elif color=='Red' or color=='Champagne' or color=='Sable':
        return 33
    else:
        return 35
    
dog_df.loc[:,'Age']=dog_df['animalcolour'].apply(age_categorized)
```

```python
dog_df['Age'].value_counts()
```

```python
dog_df['Age'].hist(color='purple',bins=8,grid=True)
plt.show()
```

```python
avg_age_by_breed = dog_df.groupby('animalbreed')['Age'].mean()
```

```python
avg_age_by_breed.plot(kind='bar', title='Mean Weight By Dog Breed')
plt.show()
```
