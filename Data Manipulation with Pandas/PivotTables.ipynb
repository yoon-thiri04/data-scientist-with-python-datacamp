```python
import pandas as pd
import numpy as np
dogs=pd.read_csv('dog.csv')
dogs.head(4)
```

```python
dogs['weight']=dogs['animaldesexed'].apply(lambda x: 40 if x=='N' else 50 if x=='Y' else 0)
```

```python
dogs_new=dogs[:30]
```

```python
dogs_new.groupby('animalcolour')['weight'].mean()
```

```python
dogs_new.pivot_table(values='weight',index='animalcolour')
```

```python
# multiple stats
dogs_new.pivot_table(values='weight',index='animalcolour',aggfunc=[np.median,np.min])
```

```python
dogs_new.groupby(['animalcolour', 'animaldesexed'])['weight'].mean()
```

```python
#Fill NaN values with 0
dogs_new.pivot_table(values='weight',index='animalcolour',columns='animaldesexed',fill_value=0)
```

```python
# Summing with pivot tables
# margins=Ture, contain the mean 
dogs_new.pivot_table(values='weight',index='animalcolour',columns='animaldesexed', fill_value=0, margins=True)
```

```python
def range(col):
    return col.quantile(0.95)-col.quantile(0.25)
```
