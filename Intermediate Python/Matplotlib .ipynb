```python
import matplotlib.pyplot as plt
import numpy as np
year=[2000,2050,2100,2150]
pop=[3.519,3.621,5.632,6.72]
year=[1800,1850,1900,1950]+year
pop=[1.0,1.56,1.78,2.1]+pop

np_pop=np.array(pop)
np_pop=np_pop*100
dict = {
    'Asia':'red',
    'Europe':'green',
    'Africa':'blue',
    'Americas':'yellow',
    'Oceania':'black'
}
plt.scatter(year,pop,c='purple',s=np_pop,alpha=0.5)
plt.ylabel('Life Expentancy[in Year]')
plt.xlabel('GDP per capita[in USD]')
#plt.xscale('log')
plt.title('Population Projection')
plt.yticks([0,2,4,6,8,10],
           ['0','2B','4B','6B','8B','10B']
          )
plt.xticks([1800,1900,2000,2100],['18k','19k','20k','21k'])
plt.grid(True)
plt.text(1880,2,'India')
plt.text(2100,4,'China')
plt.show()
```

```python
#Sub Plotting
import random
import matplotlib.pyplot as plt
x1=[]
y1=[]
for i in range(1,21):
    i=random.randint(1,50)
    x1.append(i)
for m in range(1,21):
    m=random.randint(1,30)
    y1.append(m)

x2=[]
y2=[]
for a in range(1,31):
    a=random.randint(1,60)
    x2.append(a)
for b in range(1,31):
    b=random.randint(1,50)
    y2.append(b)
    
np_y1=np.array(y1)
np_y1=np_y1*30

np_y2=np.array(y2)
np_y2=np_y2*30
fig,axes=plt.subplots(1,2)

axes[0].scatter(x1,y1,label='Plot-1',marker='o',color='blue',linestyle='-.',alpha=0.8,s=np_y1)
axes[0].set_title('Plot-1')
axes[0].set_xlabel('x_1 value')
axes[0].set_ylabel('y_1 value')
axes[0].set_xticks([0,10,20,30,40,50])
axes[0].set_yticks([0,5,10,15,20,25,30])
axes[0].grid(True)
axes[0].text(20,20,'20x20')
axes[0].legend(['Legend1'])
axes[0].annotate('Annotation1', xy=(10, 10), xytext=(15, 15), arrowprops={'facecolor':'green','shrink':0.05})

axes[1].scatter(x2,y2,label='Plot-2',marker='<',color='orange',linestyle='--',alpha=0.8,s=np_y2)
axes[1].set_title('Plot-2')
axes[1].set_xlabel('x_2 value')
axes[1].set_ylabel('y_2 value')
axes[1].set_xticks([0,10,20,30,40,50,60])
axes[1].set_yticks([0,5,10,15,20,25,30])
axes[1].text(25,25,'25x25')
axes[1].grid(True)
axes[1].legend(['Legend2'])
axes[1].annotate('Annotation2',xy=(20,15),xytext=(25,20),arrowprops={'facecolor':'yellow','shrink':0.05})

fig.legend(['Overall Legend'],loc='upper right')
plt.suptitle('Example')
plt.tight_layout()
plt.show()
plt.clear()
```

```python
x1= np.random.uniform(1.0,10.0,size=100)
x2=[]
y2=[]
for a in range(1,11):
    a=random.randint(1,20)
    x2.append(a)
for b in range(1,11):
    b=random.randint(1,20)
    y2.append(b)
np_y2=np.array(y2)
np_y2=np_y2*20
    
x3= np.random.uniform(1.0,20.0, size=100)


fig,axes=plt.subplots(2,2)
fig.legend(['Overall Legend'],loc='upper left')

axes[0,0].set_xticks([0,4,8,12,16,20,24])
#axes[0,0].set_yticks([0,2,4,6])
axes[0,0].set_title('Axis-1 Histogram')
axes[0,0].legend(['Legend1'],loc='upper left')
axes[0,0].hist(x1,bins=10,edgecolor='green',facecolor='pink')
axes[0,0].annotate('anno1',xy=(2,4), xytext=(3,7) , arrowprops={'facecolor':'red','shrink':0.07})
axes[0,0].set_xlabel('Value')
axes[0,0].set_ylabel('Frequency')
axes[0,0].grid(True)

axes[0,1].set_xticks([0,4,8,12,16,20,24])
#axes[0,1].set_yticks([0,2,4,6,8,10])
axes[0,1].set_title('Axis-2 Line Plot')
axes[0,1].legend(['Legend2'],loc='upper left')
axes[0,1].plot(x2,y2,color='blue',marker='o',linestyle='-.',alpha=0.8)
axes[0,1].annotate('anno2',xy=(6,8),xytext=(6,9),arrowprops={'facecolor':'brown','shrink':0.05})
axes[0,1].grid(True)

axes[1,0].set_xticks([0,4,8,12,16,20,24])
#axes[1,0].set_yticks([0,2,4,6,8,10])
axes[1,0].set_title('Axis-3 Scattter Plot')
axes[1,0].legend(['Legend3'],loc='upper left')
axes[1,0].scatter(x2,y2,c='yellow',marker='o',alpha=0.8,s=np_y2)
axes[1,0].annotate('anno3',xy=(5,8),xytext=(4,7),arrowprops={'facecolor':'orange','shrink':0.06})
axes[1,0].grid(True)

axes[1,1].set_xticks([0,4,8,12,16,20,24])
#axes[1,1].set_yticks([0,2,4,6,8])
axes[1,1].set_title('Axis-4 Histogram')
axes[1,1].legend(['Legend4'],loc='upper left')
axes[1,1].hist(x3,bins=20,facecolor='purple',edgecolor='yellow')
axes[1,1].annotate('anno4',xy=(5,8),xytext=(5,9),arrowprops={'facecolor':'pink','shrink':0.06})
axes[1,1].set_xlabel('Value')
axes[1,1].set_ylabel('Frequency')
axes[1,1].grid(True)

plt.tight_layout()
plt.suptitle('Four Subplots')
plt.show()
```
